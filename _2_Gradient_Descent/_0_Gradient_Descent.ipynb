```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples=4,
                      n_features=1,
                      n_informative=1,
                      n_targets = 1,
                      noise=80,
                      random_state=13)
            
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

### Let's apply OLS

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
plt.show()
```

#### Iteration 1

```python
# Let's apply Gradient Descent assuming slope is constant (m = 78.35)
# and let's assume the starting value for intercept (b=0)
m = 78.35
b = 0
y_pred = ((m*X)+b)
y_pred
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='green',label='b=0')
plt.legend()
plt.show()
```

```python
m = 78.35
b = 0

#loss_slope = -2*np.sum(y-m*X.reshape(4)- b)
loss_slope = -2*np.sum(y-m*X.ravel()- b)
loss_slope
```

```python
# Let's take learning rate = 0.1

lr = 0.1
step_size = loss_slope * lr
step_size
```

```python
# Calculating the new intercept 
b = b - step_size
b
```

```python
y_pred1 = ((m*X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1, color='#00a65a', label=f'b = {b}')
plt.plot(X,y_pred, color='#A3E4D7',label=f'b = 0')
plt.legend()
plt.show()
```

#### Iteration 2

```python
loss_slope = -2 * np.sum(y-m*X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((m*X)+ b).ravel()

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred2,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred1,color='#A3E4D7',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

#### Iteration 3

```python
loss_slope = -2*np.sum(y-m*X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope * lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred3 = ((78.35 * X) + b).ravel()

plt.figure(figsize=(15,15))
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred3,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred2,color='#A3E4D7',label='b = {}'.format(b))
plt.plot(X,y_pred1,color='#A3E4D7',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

### Using Loops
